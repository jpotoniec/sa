```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([(0,0),(0,1),(1,0),(1,1)])
y = np.array([0,0,0,1])
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.savefig('and.pdf', dpi=300)
```

```python
X = np.array([(0,0),(0,1),(1,0),(1,1)])
y = np.array([0,1,1,0])
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.savefig('xor.pdf', dpi=300)
```
