```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
feature_names = iris["feature_names"][2:]
X = iris["data"][:, 2:]  
y = (iris["target"] == 2).astype(np.int)  # 1 if Iris-Virginica, else 0
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.scatter(X[y==0,0],X[y==0,1])
plt.scatter(X[y==1,0],X[y==1,1])
plt.savefig('iris-simplified.pdf', dpi=300)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
import itertools
model = LinearRegression()
model.fit(X, y)
X_tmp = np.array(list(itertools.product(np.arange(0,7.1,.1),np.arange(0,2.6,.1))))
y_tmp = model.predict(X_tmp)
#y_tmp_cl = y_tmp>.5
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.scatter(X_tmp[:,0],X_tmp[:,1],c=y_tmp,cmap='gray')
X_tmp = np.arange(0,7.1,.1)
y_tmp = (.5-model.coef_[0]*X_tmp-model.intercept_)/model.coef_[1]
plt.plot(X_tmp, y_tmp)
#plt.scatter(X_tmp[~y_tmp_cl,0],X_tmp[~y_tmp_cl,1])
#plt.scatter(X_tmp[y_tmp_cl,0],X_tmp[y_tmp_cl,1])
plt.scatter(X[y==0,0],X[y==0,1])
plt.scatter(X[y==1,0],X[y==1,1])
plt.savefig('iris-simplified-linreg.pdf', dpi=300)
```

```python
from sklearn.linear_model import LogisticRegression
import itertools
model = LogisticRegression()
model.fit(X, y)
X_tmp = np.array(list(itertools.product(np.arange(0,7.1,.1),np.arange(0,2.6,.1))))
y_tmp = model.predict_proba(X_tmp)
#y_tmp_cl = y_tmp>.5
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.scatter(X_tmp[:,0],X_tmp[:,1],c=y_tmp[:,1],cmap='gray')
#X_tmp = np.arange(0,7.1,.1)
#y_tmp = (.5-model.coef_[0]*X_tmp-model.intercept_)/model.coef_[1]
#plt.plot(X_tmp, y_tmp)
#plt.scatter(X_tmp[~y_tmp_cl,0],X_tmp[~y_tmp_cl,1])
#plt.scatter(X_tmp[y_tmp_cl,0],X_tmp[y_tmp_cl,1])
plt.scatter(X[y==0,0],X[y==0,1])
plt.scatter(X[y==1,0],X[y==1,1])
plt.savefig('iris-simplified-logreg.pdf', dpi=300)
```

```python
feature_names = iris["feature_names"][3:]
target = iris["target_names"][2]
X = iris["data"][:, 3:]  
y = (iris["target"] == 2).astype(np.int)  # 1 if Iris-Virginica, else 0
print(X.shape, y.shape)
plt.ylabel("Prawdopodobiestwo: "+target)
plt.xlabel(feature_names[0])
plt.scatter(X[y==0],y[y==0])
plt.scatter(X[y!=0],y[y!=0])
plt.savefig('iris-simplified.pdf', dpi=300)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
plt.scatter(X[y==0],y[y==0])
plt.scatter(X[y!=0],y[y!=0])
plt.plot(X, y_pred)
plt.savefig('iris-simplified-linreg.pdf', dpi=300)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, y)
y_pred = model.predict_proba(X)
plt.scatter(X[y==0],y[y==0])
plt.scatter(X[y!=0],y[y!=0])
plt.scatter(X, y_pred[:,1])
plt.savefig('iris-simplified-logreg.pdf', dpi=300)
```
